## ðŸŽ¯ Model Details

- **Architecture**: Bayesian Hybrid Neural Network
- **Parameters**: 3.0M
- **Encoders**: 3-layer CNNs for protein and ligand
- **Uncertainty**: Bayesian variational inference
- **Ensemble**: RF + GB + DTBoost consensus

## ðŸ’¡ Tips

1. **Longer sequences**: The model handles sequences up to 1000 AA
2. **Uncertainty samples**: More samples (n_samples) = better uncertainty estimates but slower
3. **Confidence threshold**: Consider predictions with confidence > 0.7 as reliable
4. **k_off predictions**: Require molecular features and ensemble models
